EXPLORATORY DATA ANALYSIS ON TERRORISM DATASET

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

IMPORTING DATASET

```python
df = pd.read_csv('globalterrorismdb_0718dist.csv',encoding = 'latin1')
```

CLEANING OF DATA

```python
df = df[['iyear', 'imonth', 'iday','country_txt','region_txt','provstate', 'city', 'latitude', 'longitude','location', 'summary','attacktype1_txt', 'targtype1_txt','gname','motive','weaptype1_txt','nkill','nwound','addnotes']]
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df.columns
```

```python
df.rename(columns = {'iyear':"year", 'imonth':'month', 'iday':'day', 'country_txt':'country', 'region_txt':'region', 'provstate':'state',
       'city':'city', 'latitude':'latitude', 'longitude':'longitude', 'location':'location', 'summary':'summary',
       'attacktype1_txt':'attack type', 'targtype1_txt':'target type', 'gname':'group name', 'motive':'motive', 'weaptype1_txt':'weapon type',
       'nkill':'killed', 'nwound':'wounded', 'addnotes':'add notes'},inplace=True)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['killed']=df['killed'].fillna(0)
df['wounded']=df['wounded'].fillna(0)
df['casualty']=df['killed']+df['wounded']
```

```python
df.describe()
```

VISUALIZING THE DATA

1. Number of attacks in each year

```python
attacks = df['year'].value_counts(dropna=False).sort_index().to_frame().reset_index().rename(columns={'index':'year',"year":'attacks'}).set_index('year')
attacks.head()
```

```python
attacks.plot(kind='bar',color="red",figsize=(15,6),fontsize=13)
plt.title("Timeline of attacks",fontsize=15)
plt.xlabel('years',fontsize=15)
plt.ylabel('number of attacks',fontsize=15)
plt.show()
```

```python
yc = df[['year','casualty']].groupby('year').sum()
yc.head()
```

```python
yc.plot(kind = "bar",color="red",figsize=(15,6))
plt.title("year wise casualties",fontsize=15)
plt.show()
```

```python
yk = df[['year','killed']].groupby('year').sum()
yk.head()
```

```python
yk.plot(kind = "bar",color="red",figsize=(15,6))
plt.title("year wise killed",fontsize=15)
plt.show()
```

```python
yw = df[['year','wounded']].groupby('year').sum()
yw.head()
yw.plot(kind = "bar",color="red",figsize=(15,6))
plt.title("year wise wounded",fontsize=15)
plt.show()
```

2. Number of attacks in each region

```python
reg = pd.crosstab(df.year,df.region)
reg.head()
```

```python
reg.plot(kind='area',stacked=False,alpha=0.5,figsize=(20,10))
plt.title("Region wise attaks",fontsize=15)
plt.xlabel('years',fontsize=15)
plt.ylabel('number of attacks',fontsize=15)
plt.show()
```

```python
regt = reg.transpose()
regt['total']= regt.sum(axis=1)
ra=regt['total'].sort_values(ascending=False)
ra
```

```python
ra.plot(kind='bar',figsize=(15,6))
plt.title("total attacks in each region")
plt.xlabel("region")
plt.ylabel("number of attacks")
plt.show()
```

```python
rc = df[['region','casualty']].groupby('region').sum().sort_values(by='casualty',ascending=False)
rc.head()
```

```python
rc.plot(kind='bar',figsize=(15,6))
plt.title("region wise casulty")
plt.xlabel("region")
plt.ylabel("number of attacks")
plt.show()
```

```python
rk = df[['region','killed']].groupby('region').sum().sort_values(by='killed',ascending=False)
rk.head()
```

```python
rw = df[['region','wounded']].groupby('region').sum().sort_values(by='wounded',ascending=False)
rw.head()
```

```python
fig = plt.figure()
ax0 = fig.add_subplot(1,2,1)
ax1 = fig.add_subplot(1,2,2)
#killed
rk.plot(kind='bar',figsize=(15,6),ax=ax0)
ax0.set_title("people killed in each region ")
ax0.set_xlabel("region")
ax0.set_ylabel("number of kills")
#wounded
rw.plot(kind='bar',figsize=(15,6),ax=ax1)
plt.title("people wounded in each region ")
plt.xlabel("region")
plt.ylabel("number of wounded")
plt.show()
```
